#### Zadanie 4

Wyświetl wykres uwzględniając 10 województw (bez sumarycznego dla całego kraju), w których wydatki budżetów były największe w roku 2008.

> Podpowiedź: Możesz w tym celu wykorzystać moduł matplotlib.pyplot. Przykład:
```python

plt.bar([1, 2, 3], [5, 2, 4])
```

> Podpowiedź 2: Sprawdź inne funkcje aby utworzyć tytuł wykresu, oznaczyć osie, wyświetlić legendę i zmienić kolory słupków. https://matplotlib.org/stable/tutorials/introductory/pyplot.html

````python
import matplotlib.pyplot as plt
import numpy as np

with open('./dataset/dane.npz', 'rb') as f:
    content = np.load(f)
    columns, data = content['columns'].tolist(), content['data'].tolist()
    
year_column_index = columns.index('Rok')
name_column_index = columns.index('Nazwa')
filtered_data = []

for row in data:
    if row[year_column_index] == '2008' and row[name_column_index] != 'POLSKA':
        filtered_data.append(row)

value_column_index = columns.index('Wartosc')
sorted_data = sorted(filtered_data, key=lambda x: float(x[value_column_index]), reverse=True)

final_data = sorted_data[:10]

names = [row[name_column_index] for row in final_data]
values = [float(row[value_column_index]) for row in final_data]

plt.figure(figsize=(14,7))
plt.bar(names, values, color='skyblue')
plt.xlabel('Województwa', fontsize=10)
plt.ylabel('Wydatki (zł)', fontsize=10)
plt.title('10 województw o największych wydatkach budżetowych w 2008 roku')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

#### Zadanie 5

Wczytaj dane z pliku `dane-ludnosc.npz`. Plik ten zawiera zbiór danych zawierający **liczbę ludnosci**.

Oblicz wydatki z budżetu na mieszkańca dla każdego z województw (_uwzględnij dane z roku 2008_).


````python
import numpy as np

with open('./dataset/dane.npz', 'rb') as f:
    budget_data = np.load(f)
    budget_columns = budget_data['columns'].tolist()
    budget_rows = budget_data['data'].tolist()

with open('./dataset/dane-ludnosc.npz', 'rb') as f:
    population_data = np.load(f)
    population_columns = population_data['columns'].tolist()
    population_rows = population_data['data'].tolist()

budget_year_index = budget_columns.index('Rok')
budget_name_index = budget_columns.index('Nazwa')
budget_value_index = budget_columns.index('Wartosc')

population_year_index = population_columns.index('Rok')
population_name_index = population_columns.index('Nazwa')
population_value_index = population_columns.index('Wartosc')

filtered_budget_data = []
for row in budget_rows:
    if row[budget_year_index] == '2008' and row[budget_name_index] != 'POLSKA':
        filtered_budget_data.append(row)

filtered_population_data = []
for row in population_rows:
    if row[population_year_index] == '2008' and row[population_name_index] != 'POLSKA':
        filtered_population_data.append(row)

result = []
for budget_row in filtered_budget_data:
    for population_row in filtered_population_data:
        if budget_row[budget_name_index] == population_row[population_name_index]:
            budget_value = float(budget_row[budget_value_index])
            population_value = float(population_row[population_value_index])
            per_capita = budget_value / population_value
            result.append(
                {
                    'Wojewodztwo': budget_row[budget_name_index],
                    'Wydatki na mieszkanca': per_capita
                }
            )
            break

result = sorted(result, key=lambda x: x['Wydatki na mieszkanca'], reverse=True)
            
for entry in result:
    print(f"{entry['Wojewodztwo']}: {entry['Wydatki na mieszkanca']:.2f} zł/osoba")      
````

#### Zadanie 6

Które województwo odnotowało największy (oraz najmniejszy) wzrost wydatków z budżetu na mieszkańca pomiędzy latami 2002 a 2020?


````python
import numpy as np

with open('./dataset/dane.npz', 'rb') as f:
    budget_data = np.load(f)
    budget_columns = budget_data['columns'].tolist()
    budget_rows = budget_data['data'].tolist()

with open('./dataset/dane-ludnosc.npz', 'rb') as f:
    population_data = np.load(f)
    population_columns = population_data['columns'].tolist()
    population_rows = population_data['data'].tolist()

budget_year_index = budget_columns.index('Rok')
budget_name_index = budget_columns.index('Nazwa')
budget_value_index = budget_columns.index('Wartosc')

population_year_index = population_columns.index('Rok')
population_name_index = population_columns.index('Nazwa')
population_value_index = population_columns.index('Wartosc')

filtered_budget_data2002 = []
for row in budget_rows:
    if row[budget_year_index] == '2002' and row[budget_name_index] != 'POLSKA':
        filtered_budget_data2002.append(row)

filtered_population_data2002 = []
for row in population_rows:
    if row[population_year_index] == '2002' and row[population_name_index] != 'POLSKA':
        filtered_population_data2002.append(row)

result2002 = []
for budget_row in filtered_budget_data2002:
    for population_row in filtered_population_data2002:
        if budget_row[budget_name_index] == population_row[population_name_index]:
            budget_value = float(budget_row[budget_value_index])
            population_value = float(population_row[population_value_index])
            per_capita = budget_value / population_value
            result2002.append(
                {
                    'Wojewodztwo': budget_row[budget_name_index],
                    'Wydatki na mieszkanca': per_capita
                }
            )
            break

filtered_budget_data2020 = []
for row in budget_rows:
    if row[budget_year_index] == '2020' and row[budget_name_index] != 'POLSKA':
        filtered_budget_data2020.append(row)

filtered_population_data2020 = []
for row in population_rows:
    if row[population_year_index] == '2020' and row[population_name_index] != 'POLSKA':
        filtered_population_data2020.append(row)

result2020 = []
for budget_row in filtered_budget_data2020:
    for population_row in filtered_population_data2020:
        if budget_row[budget_name_index] == population_row[population_name_index]:
            budget_value = float(budget_row[budget_value_index])
            population_value = float(population_row[population_value_index])
            per_capita = budget_value / population_value
            result2020.append(
                {
                    'Wojewodztwo': budget_row[budget_name_index],
                    'Wydatki na mieszkanca': per_capita
                }
            )
            break

differences_between_years = [];
for row_02 in result2002:
    for row_20  in result2020:
        if row_02['Wojewodztwo'] == row_20['Wojewodztwo']:
            growth = row_20['Wydatki na mieszkanca'] - row_02['Wydatki na mieszkanca']
            differences_between_years.append(
                {
                    'Wojewodztwo': row_02['Wojewodztwo'],
                    'Wzrost': growth
                }
            )
            break

max_growth = max(differences_between_years, key=lambda x: x['Wzrost'])
min_growth = min(differences_between_years, key=lambda x: x['Wzrost'])  

print(f"Województwo z największym wzrostem: {max_growth['Wojewodztwo']} ({max_growth['Wzrost']:.2f} zł/osoba)")
print(f"Województwo z najmniejszym wzrostem: {min_growth['Wojewodztwo']} ({min_growth['Wzrost']:.2f} zł/osoba)")
````
